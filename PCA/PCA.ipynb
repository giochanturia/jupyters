```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.transform import Rotation as R
```

```python
Nd = 3
Np = 1000
```

```python
points = np.random.normal(0, 1, (Nd, Np))
points[0,:] *= 0.1
points = R(np.pi/3)@points
```

```python
m = points.mean(axis=1)
```

```python
plt.scatter(*points)
plt.scatter(*m)
plt.show()
```

```python
Z = points - m[:,None]
```

```python
plt.scatter(*Z)
plt.scatter(0,0)
plt.show()
```

```python
S = Z @ Z.T
```

```python
S_eigvals, S_eigvecs = np.linalg.eig(S)
```

```python
U = S_eigvecs[S_eigvals.argsort()]
```

```python
points_new = U.T @ Z
```

```python
plt.scatter(*points_new)
```
